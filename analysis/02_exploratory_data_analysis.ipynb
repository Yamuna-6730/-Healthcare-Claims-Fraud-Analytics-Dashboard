```python
import pandas as pd

import matplotlib.pyplot as plt

import plotly.express as px
```

```python
df = pd.read_csv("../dataset/health_claims_cleaned.csv")
```

```python
provider=df.groupby("Provider ID")["Billed Amount"].sum()

provider.sort_values(ascending=False).head(10)
```

```python
diagnosis=df.groupby("Diagnosis Code")["Billed Amount"].sum()

diagnosis.sort_values(ascending=False).head(10)
```

```python
monthly=df.groupby("Month")["Billed Amount"].sum()

monthly
```

```python
monthly.plot(kind="line",figsize=(10,5))
plt.title("Monthly Claim Amount")
plt.show()
```

```python
fig=px.pie(
    df,
    names="Insurance Type",
    title="Insurance Distribution"
)

fig.show()
```

```python
fig=px.bar(
    df,
    x="Claim Status",
    title="Claim Status"
)

fig.show()
```
